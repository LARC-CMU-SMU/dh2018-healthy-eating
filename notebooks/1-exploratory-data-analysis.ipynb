## 1. Loading data

Loading user profile data

```python
import pandas as pd

pro_df = pd.read_csv("../data/profiles.csv")
us_northeasts = ['CT', 'DE', 'MA', 'MD', 'ME', 'NH', 'NJ', 'NY', 'PA', 'RI', 'VT']
us_souths = ['AL', 'AR', 'FL', 'GA', 'KY', 'LA', 'MS', 'NC', 'OK', 'SC', 'TN', 'TX', 'VA', 'WV']
us_midwests = ['IA', 'IL', 'IN', 'KS', 'MI', 'MN', 'MO', 'ND', 'NE', 'OH', 'SD', 'WI']
us_wests = ['AZ', 'CA', 'CO', 'ID', 'MT', 'NM', 'NV', 'OR', 'UT', 'WA', 'WY']
us_pacifics = ['AK', 'HI']
pro_df['region'] = pro_df.apply(lambda row: 'Northeast' if row.state in us_northeasts 
                                else 'South' if row.state in us_souths 
                                else 'Midwest' if row.state in us_midwests 
                                else 'West' if row.state in us_wests 
                                else 'World', axis=1)
```

Loading diet intake data.

```python
import pandas as pd

fv_df = pd.read_csv('../data/fv_servs.csv')
prot_df = pd.read_csv('../data/protein_servs.csv')
sugar_df = pd.read_csv('../data/sugar_servs.csv')
days_df = pd.read_csv('../data/diary_days.csv')
```

```python
cal_threshold = 6000
df = days_df.groupby(["username", "date"]).sum()
df.to_csv("../data/daily_calories.csv")
```

```python
df.reset_index(inplace=True)
df = df[df.calories<=cal_threshold]
agg_df = df.groupby("username").mean().rename(columns={"calories": "mean_cal"})
agg_df = pd.merge(agg_df, df.groupby("username").median().rename(columns={"calories": "median_cal"}), 
                  left_index=True, right_index=True)
agg_df = pd.merge(agg_df, df.groupby("username").std().rename(columns={"calories": "std_cal"}), 
                  left_index=True, right_index=True)
agg_df = pd.merge(agg_df, df.groupby("username").max().rename(columns={"calories": "max_cal"}), 
                  left_index=True, right_index=True)
agg_df.to_csv("../data/cal_stats.csv")
```

Loading calories data and removing outliers (calories > mean + 2SD or negative calories).

```python
from math import ceil
import numpy as np
import pandas as pd

def week_of_month(dt):
    first_day = dt.replace(day=1)
    dom = dt.day
    adjusted_dom = dom + first_day.weekday()
    return int(ceil(adjusted_dom/7.0))

cal_df = days_df.groupby(["username", "date"]).sum()
cal_df.reset_index(inplace=True)
cal_df["date"] = pd.to_datetime(cal_df["date"])
cal_df["weekday"] = cal_df["date"].apply(lambda row: row.weekday())
cal_df["week_of_month"] = cal_df["date"].apply(lambda x: week_of_month(x))
cal_df["month"] = cal_df["date"].dt.strftime("%b")
cal_df["month_year"] = cal_df["date"].dt.strftime("%b'%y")
cal_df = pd.merge(pro_df, cal_df, how="inner", on=["username"])
```

```python
outliers = cal_df[(np.abs(cal_df.calories-cal_df.calories.mean())>(2*cal_df.calories.std())) | 
                  (cal_df.calories < 0)]
print("%% Outliers=%.4f" % (float(len(outliers))/len(cal_df)))
cal_df = cal_df[np.abs(cal_df.calories-cal_df.calories.mean())<=(2*cal_df.calories.std())]
print("# Rows after removal=%s" % len(cal_df))
```

## 3.1 Fruits and vegetables

* Figure 2(a): FV intake
* Figure 2(e): FV intake frequency

```python
import pandas as pd

fv_servs_threshold = 5

agg_df = days_df.groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_diary_days"})
df = fv_df[fv_df["num_fruit_servs"]>0].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_fruit_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = fv_df[fv_df["num_veg_servs"]>0].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_veg_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = fv_df[fv_df["num_fruit_veg_servs"]>0].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_fruit_veg_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = fv_df[fv_df["num_fruit_veg_servs"]>=fv_servs_threshold].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_sufficient_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
agg_df["pct_fruit_days"] = agg_df["num_fruit_days"]/agg_df["num_diary_days"]
agg_df["pct_veg_days"] = agg_df["num_veg_days"]/agg_df["num_diary_days"]
agg_df["pct_fruit_veg_days"] = agg_df["num_fruit_veg_days"]/agg_df["num_diary_days"]
agg_df["pct_sufficient_days"] = agg_df["num_sufficient_days"]/agg_df["num_diary_days"]
df = fv_df.groupby("username").sum()[["num_fruit_servs", "num_veg_servs", "num_fruit_veg_servs"]]
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = fv_df.groupby("username").mean()[["num_fruit_servs", "num_veg_servs", "num_fruit_veg_servs"]].rename(columns={"num_fruit_servs": "mean_fruit_servs", 
                                                            "num_veg_servs": "mean_veg_servs", 
                                                            "num_fruit_veg_servs": "mean_fruit_veg_servs"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = fv_df.groupby("username").std()[["num_fruit_servs", "num_veg_servs", "num_fruit_veg_servs"]].rename(columns={"num_fruit_servs": "std_fruit_servs", 
                                                            "num_veg_servs": "std_veg_servs", 
                                                            "num_fruit_veg_servs": "std_fruit_veg_servs"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = fv_df.groupby("username").median()[["num_fruit_servs", "num_veg_servs", "num_fruit_veg_servs"]].rename(columns={"num_fruit_servs": "median_fruit_servs", 
                                                            "num_veg_servs": "median_veg_servs", 
                                                            "num_fruit_veg_servs": "median_fruit_veg_servs"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = fv_df.groupby("username").max()[["num_fruit_servs", "num_veg_servs", "num_fruit_veg_servs"]].rename(columns={"num_fruit_servs": "max_fruit_servs", 
                                                            "num_veg_servs": "max_veg_servs", 
                                                            "num_fruit_veg_servs": "max_fruit_veg_servs"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
agg_df.fillna(0, inplace=True)
agg_df.to_csv("../data/fv_stats.csv")
```

```python
%matplotlib inline
# %matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.reset_orig() # reset seaborn's style

ls = ['-', '--', '-.', ':']

##### Mean daily fruits and vegetables servings #####
x = agg_df[(agg_df.num_diary_days>=7) & (agg_df.mean_fruit_veg_servs<=12)]["mean_fruit_veg_servs"] # select mean_fruit_veg_servs where num_diary_days >= 7
fig, ax = plt.subplots(figsize=paper_figsize)
ax.hist(x, bins=100, density=1, histtype='step', cumulative=True, label=None)
ax.grid(linestyle="dotted")
ax.set_xlabel('Mean daily servings')
ax.set_ylabel('CDF')
ax.set_xticks(range(0,int(x.max()),2))
ax.set_xlim([0,x.max()])
ax.set_ylim([0,1])
plt.tight_layout()
plt.savefig('../reports/figures/cdf_mean_fv_servs.png', dpi=200, transparent=False)
plt.show()
#####

##### % days with fruits and vegs #####
Y_labels = ["1+ servings", "5+ servings"]
Y = ["pct_fruit_veg_days", "pct_sufficient_days"]
fig, ax = plt.subplots(figsize=paper_figsize)

ls=('solid','dotted', 'dashed')
lw=(1, 1.3, 1.3)
for i, y in enumerate(Y):
    x = agg_df[agg_df['num_diary_days']>=7][y] # select pct_sufficient_days where num_diary_days >= 7
    ax.hist(x, bins=100, density=1, histtype='step', cumulative=True, label=Y_labels[i], linewidth=lw[i], linestyle=ls[i])
ax.grid(linestyle="dotted")
ax.legend(loc="lower right")
ax.set_xlabel('% Days')
ax.set_ylabel('CDF')
ax.set_xlim([0,0.95])
ax.set_ylim([0,1])
plt.tight_layout()
plt.savefig('../reports/figures/cdf_pct_fv_days.png', dpi=200, transparent=False)
plt.show()
#####
```

```python
import pandas as pd

agg_df = days_df.groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_diary_days"})

df = sugar_df[(sugar_df.num_sugar_g>0)].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_sugar_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df[(sugar_df.num_sugar_drink_ml>0)].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_sugar_drink_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df[(sugar_df.num_soft_drink_ml>0)].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_soft_drink_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df[(sugar_df.num_sugar_g>0) & (sugar_df.num_sugar_g<=25)].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_sugar_limit_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df[(sugar_df.num_sugar_drink_ml>0) & (sugar_df.num_sugar_drink_ml<=237)].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_sugar_drink_limit_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df[(sugar_df.num_soft_drink_ml>0) & (sugar_df.num_soft_drink_ml<=237)].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_soft_drink_limit_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df[sugar_df["num_dessert_g"]>0].groupby("username").agg({"date": pd.Series.nunique}).rename(columns={"date": "num_dessert_days"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)

agg_df["pct_sugar_days"] = agg_df["num_sugar_days"]/agg_df["num_diary_days"]
agg_df["pct_sugar_drink_days"] = agg_df["num_sugar_drink_days"]/agg_df["num_diary_days"]
agg_df["pct_soft_drink_days"] = agg_df["num_soft_drink_days"]/agg_df["num_diary_days"]
agg_df["pct_sugar_limit_days"] = agg_df["num_sugar_limit_days"]/agg_df["num_diary_days"]
agg_df["pct_sugar_drink_limit_days"] = agg_df["num_sugar_drink_limit_days"]/agg_df["num_diary_days"]
agg_df["pct_soft_drink_limit_days"] = agg_df["num_soft_drink_limit_days"]/agg_df["num_diary_days"]
agg_df["pct_sugar_drink_limit_rel"] = agg_df["num_sugar_drink_limit_days"]/agg_df["num_sugar_drink_days"]
agg_df["pct_soft_drink_limit_rel"] = agg_df["num_soft_drink_limit_days"]/agg_df["num_soft_drink_days"]
agg_df["pct_dessert_days"] = agg_df["num_dessert_days"]/agg_df["num_diary_days"]

df = sugar_df.groupby("username").sum()[["num_sugar_g", "num_sugar_drink_ml", "num_soft_drink_ml", "num_dessert_g"]]
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df.groupby("username").mean()[["num_sugar_g", "num_sugar_drink_ml", "num_soft_drink_ml", "num_dessert_g"]].rename(columns={"num_sugar_g": "mean_sugar_g", 
                                                                                                                                      "num_sugar_drink_ml": "mean_sugar_drink_ml", 
                                                                                                                                    "num_soft_drink_ml": "mean_soft_drink_ml", 
                                                                                                                                    "num_dessert_g": "mean_dessert_g"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df.groupby("username").std()[["num_sugar_g", "num_sugar_drink_ml", "num_soft_drink_ml", "num_dessert_g"]].rename(columns={"num_sugar_g": "std_sugar_g", 
                                                                                                                                    "num_sugar_drink_ml": "std_sugar_drink_ml", 
                                                                                                                                    "num_soft_drink_ml": "std_soft_drink_ml", 
                                                                                                                                    "num_dessert_g": "std_dessert_g"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df.groupby("username").median()[["num_sugar_g", "num_sugar_drink_ml", "num_soft_drink_ml", "num_dessert_g"]].rename(columns={"num_sugar_g": "median_sugar_g", 
                                                                                                                                        "num_sugar_drink_ml": "median_sugar_drink_ml", 
                                                                                                                                        "num_soft_drink_ml": "median_soft_drink_ml", 
                                                                                                                                        "num_dessert_g": "median_dessert_g"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
df = sugar_df.groupby("username").max()[["num_sugar_g", "num_sugar_drink_ml", "num_soft_drink_ml", "num_dessert_g"]].rename(columns={"num_sugar_g": "max_sugar_g", 
                                                                                                                                        "num_sugar_drink_ml": "max_sugar_drink_ml", 
                                                                                                                                        "num_soft_drink_ml": "max_soft_drink_ml", 
                                                                                                                                        "num_dessert_g": "max_dessert_g"})
agg_df = pd.merge(agg_df, df, how="outer", left_index=True, right_index=True)
agg_df.fillna(0, inplace=True)
agg_df.to_csv("../data/sugar_stats.csv")
```
