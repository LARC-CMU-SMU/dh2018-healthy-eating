```python
import numpy as np
import pandas as pd

def label_age_group(age):
    if age >= 18 and age <= 44:
        return "18-44"
    elif age>=45:
        return "45+"

def label_age_generation(age):
    if age >= 18 and age <= 34:
        return "Millennials"
    elif age >= 35 and age <= 50:
        return "Gen X"
    elif age >= 51 and age <= 100:
        return "Baby boomers"
    
def label_friends_group(num_friends):
    if num_friends >= 0 and num_friends <= 6:
        return "Q1"
    elif num_friends > 6 and num_friends <= 18:
        return "Q2"
    elif num_friends > 18 and num_friends <= 41:
        return "Q3"
    elif num_friends > 41:
        return "Q4"

def label_regions_us(state):
    if state in us_northeasts:
        return "Northeast"
    elif state in us_souths:
        return "South"
    elif state in us_midwests:
        return "Midwest"
    elif state in us_wests:
        return "West"

def label_regions_all(state):
    if state in us_northeasts+us_souths+us_midwests+us_wests:
        return "US"
    else:
        return "Non-US"

pro_df["age_group"] = pro_df.age.apply(lambda x: label_age_group(x))
pro_df["age_generation"] = pro_df.age.apply(lambda x: label_age_generation(x))
pro_df["friends_group"] = pro_df.friends_count.apply(lambda x: label_friends_group(x))
pro_df["regions_us"] = pro_df.state.apply(lambda x: label_regions_us(x))
pro_df["regions_all"] = pro_df.state.apply(lambda x: label_regions_all(x))

df = pro_df.groupby("gender").agg({"id": pd.Series.nunique})[["id"]].rename(columns={"id": "count"}).sort_values(by="count", ascending=False)
df["pct"] = df["count"]/sum(df["count"])
df2 = pro_df.groupby("age_group").agg({"id": pd.Series.nunique})[["id"]].rename(columns={"id": "count"}).sort_values(by="count", ascending=False)
df2["pct"] = df2["count"]/sum(df2["count"])
df = df.append(df2)
df2 = pro_df.groupby("age_generation").agg({"id": pd.Series.nunique})[["id"]].rename(columns={"id": "count"}).sort_values(by="count", ascending=False)
df2["pct"] = df2["count"]/sum(df2["count"])
df = df.append(df2)
df2 = pro_df.groupby("friends_group").agg({"id": pd.Series.nunique})[["id"]].rename(columns={"id": "count"}).sort_values(by="count", ascending=False)
df2["pct"] = df2["count"]/sum(df2["count"])
df = df.append(df2)
df2 = pro_df.groupby("regions_us").agg({"id": pd.Series.nunique})[["id"]].rename(columns={"id": "count"}).sort_values(by="count", ascending=False)
df2["pct"] = df2["count"]/sum(df2["count"])
df = df.append(df2)
df2 = pro_df.groupby("regions_all").agg({"id": pd.Series.nunique})[["id"]].rename(columns={"id": "count"}).sort_values(by="count", ascending=False)
df2["pct"] = df2["count"]/sum(df2["count"])
df = df.append(df2)
```

## 2. Statistical Tests

To measure the differences in eating behaviors of food journalers across sociodemographic groups, we are going to perform:

* __Mann-Whitney U-test__ for genders, age groups, and global regions.
* __Kruskal-Wallis H-test with Dunn's multiple comparison test__ for social connections and US regions.

### 2. 1 Preparing data and setting up hypothesis testings

```python
import numpy as np
import pandas as pd
from scipy import stats
# workaround for the stats.chisqprob error (https://github.com/statsmodels/statsmodels/issues/3931)
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
from itertools import combinations
from statsmodels.stats.multitest import multipletests
from statsmodels.stats.libqsturng import psturng
import warnings


def kw_dunn(groups, to_compare=None, alpha=0.05, method='bonf'):
    """

    Kruskal-Wallis 1-way ANOVA with Dunn's multiple comparison test.
    Implemented by https://gist.github.com/alimuldal/fbb19b73fa25423f02e8

    Arguments:
    ---------------
    groups: sequence
        arrays corresponding to k mutually independent samples from
        continuous populations

    to_compare: sequence
        tuples specifying the indices of pairs of groups to compare, e.g.
        [(0, 1), (0, 2)] would compare group 0 with 1 & 2. by default, all
        possible pairwise comparisons between groups are performed.

    alpha: float
        family-wise error rate used for correcting for multiple comparisons
        (see statsmodels.stats.multitest.multipletests for details)

    method: string
        method used to adjust p-values to account for multiple corrections (see
        statsmodels.stats.multitest.multipletests for options)

    Returns:
    ---------------
    H: float
        Kruskal-Wallis H-statistic

    p_omnibus: float
        p-value corresponding to the global null hypothesis that the medians of
        the groups are all equal

    Z_pairs: float array
        Z-scores computed for the absolute difference in mean ranks for each
        pairwise comparison

    p_corrected: float array
        corrected p-values for each pairwise comparison, corresponding to the
        null hypothesis that the pair of groups has equal medians. note that
        these are only meaningful if the global null hypothesis is rejected.

    reject: bool array
        True for pairs where the null hypothesis can be rejected for the given
        alpha

    Reference:
    ---------------
    Gibbons, J. D., & Chakraborti, S. (2011). Nonparametric Statistical
    Inference (5th ed., pp. 353-357). Boca Raton, FL: Chapman & Hall.

    """

    # omnibus test (K-W ANOVA)
    # -------------------------------------------------------------------------

    groups = [np.array(gg) for gg in groups]

    k = len(groups)

    n = np.array([len(gg) for gg in groups])
    if np.any(n < 5):
        warnings.warn("Sample sizes < 5 are not recommended (K-W test assumes "
                      "a chi square distribution)")

    allgroups = np.concatenate(groups)
    N = len(allgroups)
    ranked = stats.rankdata(allgroups)

    # correction factor for ties
    T = stats.tiecorrect(ranked)
    if T == 0:
        raise ValueError('All numbers are identical in kruskal')

    # sum of ranks for each group
    j = np.insert(np.cumsum(n), 0, 0)
    R = np.empty(k, dtype=np.float)
    for ii in range(k):
        R[ii] = ranked[j[ii]:j[ii + 1]].sum()

    # the Kruskal-Wallis H-statistic
    H = (12. / (N * (N + 1.))) * ((R ** 2.) / n).sum() - 3 * (N + 1)

    # apply correction factor for ties
    H /= T

    df_omnibus = k - 1
    p_omnibus = stats.chisqprob(H, df_omnibus)

    # multiple comparisons
    # -------------------------------------------------------------------------

    # by default we compare every possible pair of groups
    if to_compare is None:
        to_compare = tuple(combinations(range(k), 2))

    ncomp = len(to_compare)

    Z_pairs = np.empty(ncomp, dtype=np.float)
    p_uncorrected = np.empty(ncomp, dtype=np.float)
    Rmean = R / n

    for pp, (ii, jj) in enumerate(to_compare):

        # standardized score
        Zij = (np.abs(Rmean[ii] - Rmean[jj]) /
               np.sqrt((1. / 12.) * N * (N + 1) * (1. / n[ii] + 1. / n[jj])))
        Z_pairs[pp] = Zij

    # corresponding p-values obtained from upper quantiles of the standard
    # normal distribution
    p_uncorrected = stats.norm.sf(Z_pairs) * 2.

    # correction for multiple comparisons
    reject, p_corrected, alphac_sidak, alphac_bonf = multipletests(
        p_uncorrected, method=method
    )

    return H, p_omnibus, Z_pairs, p_corrected, reject
```

```python
from itertools import combinations
import numpy as np
from scipy import stats

# test variables
y_groups = {
            'diary': ['num_diary_days', "pct_lapsed_days", "mean_cal", "median_cal"],
            'fruit_veg': ['num_fruit_veg_servs', 'num_fruit_veg_days', 'num_sufficient_days', 
                          'mean_fruit_veg_servs', 'median_fruit_veg_servs', 
                          'pct_fruit_veg_days', 'pct_sufficient_days'], 
            'meat': ['num_red_meat_servs', 'num_poultry_servs', 'num_fish_servs', 
                     'num_red_meat_days', 'num_poultry_days', 'num_fish_days', 
                     'mean_red_meat_servs', 'mean_poultry_servs', 'mean_fish_servs', 
                     'median_red_meat_servs', 'median_poultry_servs', 'median_fish_servs', 
                     'pct_red_meat_days', 'pct_poultry_days', 'pct_fish_days'], 
            'sugar': ['num_sugar_g', 'num_sugar_drink_ml', 'num_soft_drink_ml', 'num_dessert_g', 
                      'num_sugar_drink_days', 'num_soft_drink_days', 'num_dessert_days', 
                      'mean_sugar_g', 'mean_sugar_drink_ml', 'mean_soft_drink_ml', 'mean_dessert_g', 
                      'median_sugar_g', 'median_sugar_drink_ml', 'median_soft_drink_ml', 'median_dessert_g', 
                      'pct_sugar_days', 'pct_sugar_drink_days', 'pct_soft_drink_days', 'pct_dessert_days'],
            }

group_vars = {"gender": ['Male', 'Female'], 
              "age_group": ["18-44", "45+"], 
              "age_generation": ["Millennials", "Gen X", "Baby boomers"],
              "regions_us": ["Northeast", "South", "Midwest", "West"], 
              "regions_all": ["US", "Non-US"], 
#               "friends_group": ["0-9", "10-49", "50-99", "100-5000"]
              "friends_group": ["Q1", "Q2", "Q3", "Q4"]
             }

stat_tests = []
posthoc_tests = []
tasks = ["diary", "fruit_veg", "meat", "sugar"]
sig_level = 0.05

# For each task,
for task in tasks:
    Yi = y_groups[task]
    # For each test variable in task
    for y in Yi:
        is_normal = False
                
        # For each subgroup of data
        for grp, vs in group_vars.iteritems():
            data = []
            max_grp = None
            median_diff_pct = None
            
            for v in vs:
                data.append(df[df[grp]==v][y])      
            
            medians = [np.median(d) for d in data]
            
            if is_normal:
                if len(data) == 2:
                    test_name = "t_test"
                    res = stats.ttest_ind(*data)
                elif len(data) > 2:
                    test_name = "f_oneway"
                    res = stats.f_oneway(*data)
            else:
                if len(data) == 2:
                    test_name = "mannwhitney_u"
                    res = stats.mannwhitneyu(*data)
                    
                    if res[1] < sig_level:
                        median_diff_pct = abs(medians[0]-medians[1])/(np.min(medians))
                        max_grp = vs[medians.index(np.max(medians))]
                
                elif len(data) > 2:
                    test_name = "kruskal_h"
                    res = stats.kruskal(*data)
                    
                    if res[1] < sig_level:
                        max_grp = vs[medians.index(np.max(medians))]
                                                
                        # Posthoc
                        pairs = list(combinations(vs, 2))
                        H, p_omnibus, Z_pairs, p_corrected, reject = kw_dunn(data, to_compare=None, alpha=0.05, method='bonf')
                        sig_pairs = []
                        
                        for i, r in enumerate(reject):
                            if r:
                                sig_pairs.append(pairs[i])
                        
                        posthoc_tests.append({"task": task, "y": y, "group": grp, "H": H, "p_omnibus": p_omnibus, 
                                              "Z_pairs": Z_pairs, "p_corrected": p_corrected, "reject": reject, 
                                              "sig_pairs": sig_pairs})
            
            stat_tests.append({"task": task, "y": y, "group": grp, "test": test_name, 
                               "stat": res[0], "p-value": res[1], 
                               "max_group": max_grp, 
                               "medians": [float("%.5f" % i) for i in medians] if medians else None, 
                               "median_diff_pct": "%.5f" % median_diff_pct if median_diff_pct else None})

stats_df = pd.DataFrame.from_dict(stat_tests)
stats_df = stats_df[["task", "y", "group", "test", "stat", "p-value", "max_group", "medians", "median_diff_pct"]].sort_values(by=["task", "y", "group"])
posthoc_df = pd.DataFrame.from_dict(posthoc_tests)
posthoc_df = posthoc_df[["task", "y", "group", "H", "p_omnibus", "Z_pairs", "p_corrected", "reject", "sig_pairs"]].sort_values(by=["task", "y", "group"])
```

## 3. Regression analysis

We are going to run regression analysis to measure:
* The influences of demographic factors on the journaling behaviors.
* The influences of demographic and journaling factors on the caloric and diet-based behavioral measures.

### 3.1 Preparing data and setting up regression tasks

```python
import numpy as np
import time

# Dependent variables
y_groups = {'diary': ["num_diary_days", "pct_lapsed_days"], 
            'calorie': ["mean_cal", "median_cal"],
            'fruit_veg': ['num_fruit_veg_servs', 'num_fruit_veg_days', 'num_sufficient_days', 
                          'mean_fruit_veg_servs', 'median_fruit_veg_servs', 
                          'pct_fruit_veg_days', 'pct_sufficient_days'], 
            'meat': ['num_red_meat_servs', 'num_poultry_servs', 'num_fish_servs', 
                     'num_red_meat_days', 'num_poultry_days', 'num_fish_days', 
                     'mean_red_meat_servs', 'mean_poultry_servs', 'mean_fish_servs', 
                     'median_red_meat_servs', 'median_poultry_servs', 'median_fish_servs', 
                     'pct_red_meat_days', 'pct_poultry_days', 'pct_fish_days'], 
            'sugar': ['num_sugar_g', 'num_sugar_drink_ml', 'num_soft_drink_ml', 'num_dessert_g', 
                      'num_sugar_drink_days', 'num_soft_drink_days', 'num_dessert_days', 
                      'mean_sugar_g', 'mean_sugar_drink_ml', 'mean_soft_drink_ml', 'mean_dessert_g', 
                      'median_sugar_g', 'median_sugar_drink_ml', 'median_soft_drink_ml', 'median_dessert_g',
                      'pct_sugar_days' ,'pct_sugar_drink_days', 'pct_soft_drink_days', 'pct_dessert_days'],
            }

# Feature types
# Keys should correspond to those in y_groups
feature_groups = {"profile": ["is_male", "age", "log_friends_count", "Northeast", "South", "West", "Non-US"], 
                 "diary_behavior": ["num_diary_days", "pct_lapsed_days"]}
task_features = {'diary': feature_groups["profile"], 
                 'calorie': feature_groups["profile"]+feature_groups["diary_behavior"],
                 'fruit_veg': feature_groups["profile"]+feature_groups["diary_behavior"],
                 'meat': feature_groups["profile"]+feature_groups["diary_behavior"],
                 'sugar': feature_groups["profile"]+feature_groups["diary_behavior"],}


# Known outliers: These user data will be excluded
outliers = {'sugar': ['3936', '527', '476', '5883', '477']}
    
Ys = {}
Ds = {}

for k, y_group in y_groups.iteritems():
    if outliers and k in outliers:
        df_k = df[~df.username.isin(outliers[k])]
    else:
        df_k = df

    Y = dict()
    for y in y_group:
        Y.update({y: df_k[[y]]})
    Ys.setdefault(k, Y)
    Ds.setdefault(k, df_k[task_features[k]])
    print "%s\t%s users (rows)" % (k, len(df_k))
```

__Checking for collinearity__: Look for very small eigen values (close to zero). Seems fine.

```python
import numpy as np

tasks = ['diary', 'fruit_veg', 'meat', 'sugar']
print "Eigen values:"

for task in tasks:
    X = Ds[task]
    corr = np.corrcoef(X, rowvar=0)
    eigen_values, eigen_vector = np.linalg.eig(corr)
    print "%s\t%s" % (task, "\t".join([str("%.4f" % i) for i in eigen_values]))
```

### 3.2 Running Regression


```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import statsmodels.api as sm

def do_regression(X, y, test_size, fit_intercept, y_label, print_summary=True):
    if test_size:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=100)
    else:
        y_train = y
        X_train = X
    
    if fit_intercept:
        model = sm.OLS(y_train, sm.add_constant(X_train)).fit()
    else:
        model = sm.OLS(y_train, X_train).fit()
      
    if print_summary:
        print y_label
        print model.summary()
        print ''
    else:
        result = {"y": y_label, "rsquared": model.rsquared, "rsquared_adj": model.rsquared_adj}
        presult = {"y": y_label}
        for i in range(0, len(model.params)):
            if i > 0:
                key = "x"+str(i)
            else:
                key = "const"
            result.update({key: model.params[i]})
            presult.update({key: model.pvalues[i]})
    
    return pd.DataFrame.from_dict([result]), pd.DataFrame.from_dict([presult])

tasks = ['diary', 'calorie','fruit_veg', 'meat', 'sugar']
print_summary = False
reg_df = None
feature_names = []
itasks = []

for task in tasks:
    X = Ds[task]
    Y = Ys[task]

    if len(task_features[task]) > len(feature_names):
        feature_names = task_features[task]

    for y_label, y in Y.iteritems():
        itasks.append(task)
        
        if reg_df is None:
            reg_df, pv_df = do_regression(X, y, None, True, y_label, print_summary)
        else:
            reg_df = pd.merge(reg_df, do_regression(X, y, None, True, y_label, print_summary)[0], how="outer")
            pv_df = pd.merge(pv_df, do_regression(X, y, None, True, y_label, print_summary)[1], how="outer")

reg_df["task"] = pd.Series(itasks)
pv_df["task"] = pd.Series(itasks)

xfeats = []
for i in range(len(feature_names)):
    xfeats.append("x"+str(i+1))

reg_df = reg_df[["task", "y", "const"] + xfeats + ["rsquared", "rsquared_adj"]].sort_values(by=["task", "y"])
pv_df = pv_df[["task", "y", "const"] + xfeats].sort_values(by=["task", "y"])
for i in range(len(feature_names)):
    reg_df.rename(columns={"x"+str(i+1): feature_names[i]}, inplace=True)
    pv_df.rename(columns={"x"+str(i+1): feature_names[i]}, inplace=True)
```
